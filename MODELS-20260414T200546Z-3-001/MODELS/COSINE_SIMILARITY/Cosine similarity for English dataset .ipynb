```python

correct = 1
for i in range(50):
    p = file.context[i]
    q = file.question[i]
    a = file.text[i]
#     print("Paragraph Before Removing Stop words  "+p)
#     print("===")
#     print("===")
#     print("===")

    word_tokens = word_tokenize(p)
    filtered_sentence = [w for w in word_tokens if not w.lower() in data]
    filtered_sentence = []
    for w in word_tokens:
        if w not in eng_stop_words:
            filtered_sentence.append(w)
    #print(filtered_sentence)

    mystring = ' '.join(map(str,filtered_sentence))
#     print("My string  "+mystring)
#     print("===")
#     print("===")
#     print("===")
    splitted_sen = mystring.split(".")

#     print("===")
#     print("===")
#     print("===")
    word_tokens = word_tokenize(q)
    filtered_question = [w for w in word_tokens if not w.lower() in data]
    filtered_question = []
    for t in word_tokens:
        if t not in eng_stop_words:
            filtered_question.append(t)
    #print(filtered_question)

    string = ' '.join(map(str,filtered_question))
#     print("My string  "+string)
#     print("===")
#     print("===")
#     print("===")

    splitted_ques = string.split(".")

    csl = []
#     print(len(a_list))
    for i in (splitted_sen):
        
        cosine_similarity = nlp(i).similarity(nlp(splitted_ques[0]))
        csl.append(cosine_similarity)
#     print(csl)
#     print("===")
#     print("===")
#     print("===")



    res = {}
    for key in splitted_sen:
        for value in csl:
            res[key] = value
            csl.remove(value)
            break  

#     Printing resultant dictionary 
#     print ("Resultant dictionary is : " +  str(res))


    fin_max = max(res, key=res.get)
    # All ANSWERS
#     print("Question: " + q)
#     print("Maximum value:",fin_max)


    if a in  fin_max:
        correct += 1
        # CORRECT ANSWERS
#         print("Actual Answer"+a)
#         print("Predicted Answer"+fin_max)
            

    else:
        pass

print(correct)
# print("Accuracy in %",((correct/50)*100))

#     print(a.value, b.value,c.value,d.value)
```
